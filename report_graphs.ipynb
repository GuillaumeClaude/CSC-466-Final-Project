```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\guill\CSC-466-Final-Project\data\Test_Out.csv", sep=",")

delay = pd.read_csv(r"C:\Users\guill\CSC-466-Final-Project\data\delay.csv", sep=",")
delay = delay.drop(['Unnamed: 2'], axis = 1)


#clean ie no laundering
clean = pd.read_csv(r"C:\Users\guill\CSC-466-Final-Project\data\no_laundering.csv", sep=",")
clean = clean.drop(['Unnamed: 2'], axis = 1)

batch = pd.read_csv(r"C:\Users\guill\CSC-466-Final-Project\data\batching.csv", sep=",")
batch = batch.drop(['Unnamed: 2'], axis = 1)
batch.head()

dummy = pd.read_csv(r"C:\Users\guill\CSC-466-Final-Project\data\dummies.csv", sep=",")
dummy = dummy.drop(['Unnamed: 2'], axis = 1)
dummy.head()

constant = pd.read_csv(r"C:\Users\guill\CSC-466-Final-Project\data\constantall.csv", sep=",")
constant = constant.drop(['Unnamed: 2'], axis = 1)
constant.head()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x = delay['t']
x = x.tolist()
print("x axis", x)

y_delay = delay['detection_rate']
y_delay = y_delay.tolist()
print("delay data", y_delay)

y_clean = clean['detection_rate']
y_clean = y_clean.tolist()
print("clean data", y_clean)

y_batch = batch['detection_rate']
y_batch = y_batch.tolist()
print("batch data", y_batch)

y_dummy = dummy['detection_rate']
y_dummy = y_dummy.tolist()
print("dummy data", y_dummy)

y_constant = constant['detection_rate']
y_constant = y_constant.tolist()
print("constant data", y_constant)

plt.plot(x, y_delay, marker='o', linestyle='--', color='tab:blue', label='Delay')
plt.plot(x, y_clean, marker='x', linestyle='--', color='tab:orange', label='No laundering')
plt.plot(x, y_batch, marker='s', linestyle='--', color='tab:green', label='Batching')
plt.plot(x, y_dummy, marker='+', linestyle='--', color='tab:red', label='Dummy', markersize='7.0')
plt.plot(x, y_constant, marker='D', linestyle='--', color='tab:purple', label='Constant')

plt.legend()
plt.ylim(0,1.1)
plt.xlabel("Data Collection Time")
plt.ylabel("Detection Rate")
plt.show()
```
